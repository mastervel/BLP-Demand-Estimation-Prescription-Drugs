```python
import pandas as pd
import os
import matplotlib.pyplot as plt
import textwrap
```

```python
def abbreviate_label(label, max_length=15):
    return textwrap.shorten(label, width=max_length, placeholder='...')
```

```python
pd.set_option('display.max_rows', 100)  # Set the desired number of rows (e.g., 100)

# Increase the number of columns displayed
pd.set_option('display.max_columns', 100)  # Set the desired number of columns (e.g., 100)

# Increase the width of the output display
pd.set_option('display.width', 1000)
```

```python
folder_path = 'C:\\Users\\Veli\\Downloads\\Code\\data\\clean\\estimated'

# Initialize an empty list to store DataFrames from each file
dataframes_list = []

# Loop through files in the folder
for filename in os.listdir(folder_path):
    # Check if the file is a CSV file
    if filename.endswith('.csv'):
        # Construct the full file path
        file_path = os.path.join(folder_path, filename)
        
        # Read the CSV file into a DataFrame and append it to the list
        df = pd.read_csv(file_path)
        dataframes_list.append(df)

# Concatenate the list of DataFrames vertically into one DataFrame
merged_df = pd.concat(dataframes_list, ignore_index=True, sort=False)
```

```python
# for column in own_prices_filtered.columns:
#     plt.plot(own_prices_filtered.index, own_prices_filtered[column], label=abbreviate_label(column, 20))
plt.figure(figsize=(12, 6))

for column in own_prices_filtered_neg.columns:
    data = own_prices_filtered_neg[column].dropna()  # Drop NaN values from the column
    plt.plot(data.index, data, label=abbreviate_label(column, 60))
    
# Add legend and labels
plt.title('Own-Price Elasticities Over Time')
plt.legend(loc='upper right', bbox_to_anchor=(1, -0.1), ncol=2)
plt.xlabel('Time')
plt.ylabel('Own-Price Elasticities')

# Show the plot
plt.show()
```

```python
threshold = 5.0

# Calculate the absolute mean value for each column
abs_mean_values = own_prices_filtered_neg.abs().mean()

# Filter columns based on the threshold condition
columns_to_keep = abs_mean_values[abs_mean_values >= threshold].index

# Create a new DataFrame with the selected columns
df_filtered = own_prices_filtered_neg[columns_to_keep]
```
